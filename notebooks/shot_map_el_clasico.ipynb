```python
from mplsoccer import Pitch
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Créer le terrain
pitch = Pitch(pitch_type='statsbomb', 
              pitch_color='#1a2e1a', 
              line_color='white')

fig, ax = pitch.draw(figsize=(12, 8))
```

```python
# Séparer les tirs par équipe
shots_rm = shots[shots['team'] == 'Real Madrid']
shots_barca = shots[shots['team'] == 'Barcelona']
```

```python
# Extraire les coordonnées    
def get_coords(shots_df):
    x = [loc[0] for loc in shots_df['location']]
    y = [loc[1] for loc in shots_df['location']] 

    return x, y
```

```python
x_rm, y_rm = get_coords(shots_rm)   
x_barca, y_barca = get_coords(shots_barca)
```

```python
# Taille des cercles = xG
size_rm = shots_rm['shot_statsbomb_xg'] * 1000
size_barca = shots_barca['shot_statsbomb_xg'] * 1000
```

```python
# Couleur selon résultat
color_rm = ['#00e676' if o == 'Goal' else 'white' for o in shots_rm['shot_outcome']]
color_barca = ['#ff3d3d' if o == 'Goal' else 'white' for o in shots_barca['shot_outcome']]
```

```python
# Dessiner les tirs
pitch.scatter(x_rm, y_rm, s=size_rm, c=color_rm, 
              edgecolors='#00e676', alpha=0.8, ax=ax, label='Real Madrid')
pitch.scatter(x_barca, y_barca, s=size_barca, c=color_barca, 
              edgecolors='#ff3d3d', alpha=0.8, ax=ax, label='Barcelona')
```

```python
# Titre
ax.set_title('El Clasico — Shot Map\nBarcelona 1-3 Real Madrid · 24 Oct 2020', 
             color='white', fontsize=16, pad=20)

plt.legend(loc='upper left', facecolor='#1a2e1a', labelcolor='white')
plt.tight_layout()
plt.show()
```

```python
from mplsoccer import Pitch
import matplotlib.pyplot as plt

# Créer le terrain avec le bon système de coordonnées
pitch = Pitch(pitch_type='statsbomb',
              pitch_color='#1a2e1a',
              line_color='white')

fig, ax = pitch.draw(figsize=(12, 8))

# Séparer les tirs par équipe
shots_rm = shots[shots['team'] == 'Real Madrid']
shots_barca = shots[shots['team'] == 'Barcelona']

# Extraire les coordonnées
def get_coords(shots_df):
    x = [loc[0] for loc in shots_df['location']]
    y = [loc[1] for loc in shots_df['location']]
    return x, y

x_rm, y_rm = get_coords(shots_rm)
x_barca, y_barca = get_coords(shots_barca)

# Taille des cercles = xG
size_rm = shots_rm['shot_statsbomb_xg'] * 1000
size_barca = shots_barca['shot_statsbomb_xg'] * 1000

# Couleur selon résultat
color_rm = ['#00e676' if o == 'Goal' else 'white' 
            for o in shots_rm['shot_outcome']]
color_barca = ['#ff3d3d' if o == 'Goal' else 'white' 
               for o in shots_barca['shot_outcome']]

# Dessiner les tirs
pitch.scatter(x_rm, y_rm, s=size_rm, c=color_rm,
              edgecolors='#00e676', alpha=0.8, ax=ax)
pitch.scatter(x_barca, y_barca, s=size_barca, c=color_barca,
              edgecolors='#ff3d3d', alpha=0.8, ax=ax)

# Titre
ax.set_title('El Clasico — Shot Map\nBarcelona 1-3 Real Madrid · 24 Oct 2020',
             color='white', fontsize=16, pad=20)

plt.tight_layout()
plt.show()
```
